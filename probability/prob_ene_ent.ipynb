# Distribution of states

We have $N$ particles and £k£ energy levels.

State is defined by defining the level of each particle $k_i$:
$$\boldsymbol{S} = (k_1, k_2, k_3, ..., k_n)$$

Distribution is defined as the number of particles in each of the energy levels:
$$\boldsymbol{N} = (N_0, N_1, ..., N_k)$$

#### Number of particles
$$N = \sum_{k} N_k$$

#### Total energy
$$E = \sum_{k} N_k E_k$$


## Simple example

Three energy levels:
$E_0 = 0$; $E_1 = 1\epsilon$; $E_2 = 3\epsilon$

#### Possible distributions for
$N = 3$; $E = 3 \epsilon$

| # | $\boldsymbol{S}$ | $\boldsymbol{N}$ | 
| :-: | :-: | :-: | 
| 1 | (1,1,1) | (0,3,0) | 
| 2 | (2,0,0) | (2,0,1) | 
| 3 | (0,2,0) | (2,0,1) | 
| 4 | (0,0,2) | (2,0,1) | 

Distrubution (0,3,0) has 1 realization, while distribution (2,0,1) has 3

#### Possible distributions for
$N = 4$; $E = 3 \epsilon$

| # | $\boldsymbol{S}$ | $\boldsymbol{N}$ | 
| :-: | :-: | :-: | 
| 1 | (1,1,1,0) | (1,3,0) | 
| 2 | (1,1,0,1) | (1,3,0) | 
| 3 | (1,0,1,1) | (1,3,0) | 
| 4 | (0,1,1,1) | (1,3,0) | 
| 5 | (2,0,0,0) | (3,0,1) | 
| 6 | (0,2,0,0) | (3,0,1) | 
| 7 | (0,0,2,0) | (3,0,1) | 
| 8 | (0,0,0,2) | (3,0,1) | 

Distrubution (1,3,0) has 4 realizations, while distribution (3,0,1) has 4

#### Possible distributions for $E = 3 \epsilon$ (as a function of $N$)

$\boldsymbol{N_1} = (N - 3, 3, 0)$; $E = (N-3) \cdot 0 + 3 \cdot 1 \epsilon + 0 = 3 \epsilon$ 

$\boldsymbol{N_2} = (N - 1, 0, 1)$; $E = (N-1) \cdot 0 + 0 + 1 \cdot 3 \epsilon = 3 \epsilon$ 


Distrubution $\boldsymbol{N_1}$ ($(N - 3, 3, 0)$) has  ${N\choose 3}$ realizations, while distribution $\boldsymbol{N_2}$ ($(N - 1, 0, 1)$ has $N$




```python
import numpy as np
from matplotlib import pyplot as plt
import scipy

x = []
P1 = []
P2 = []
for N in range(3, 20):
    N_N1 = scipy.special.comb(N, 3)
    N_N2 = N
    N_sum = N_N1 + N_N2
    p1 = N_N1 / N_sum
    p2 = 1 - p1
    x.append(N)
    P1.append(p1)
    P2.append(p2)
    

labs = "${N_1}$", '${N_2}$'
plt.plot(x, P1, label=labs[0])
plt.plot(x, P2, label=labs[1])
plt.legend()
plt.xlabel('$N$')
plt.ylabel('$p$')
plt.show()
```
